```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset/Training Data.csv")

print(df.info())
print(df["Risk_Flag"].value_counts())
```

```python

import matplotlib.pyplot as plt

# Risk_Flag 數據
risk_counts = df["Risk_Flag"].value_counts()
labels = ["No Risk (0)", "Risk (1)"]
colors = ["#66b3ff", "#ff9999"]
explode = (0.05, 0.05)

# 自定義標籤格式
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count:,})'
    return my_autopct

# 畫圖
plt.figure(figsize=(6, 6))
plt.pie(risk_counts, labels=labels, autopct=make_autopct(risk_counts), 
        startangle=140, colors=colors, explode=explode)
plt.title("Risk Flag Distribution", fontsize=14, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("Dataset/Training Data.csv")

numeric_df = df.drop(columns=['Id']).select_dtypes(include=['int64', 'float64'])

# 計算皮爾森相關係數矩陣
corr_matrix = numeric_df.corr()

# 繪製相關係數熱力圖
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", square=True, cbar_kws={"shrink": 0.8})
plt.title("Risk Flag Correlation Matrix", fontsize=14)
plt.tight_layout()
plt.show()
```
